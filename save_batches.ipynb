```python
import torch 
# import yaml
import os
import os.path as osp
from tqdm import tqdm
import sys

epochs_batches = []
for epoch in tqdm(range(130), position=0):
    epoch_folder = osp.join('batch_logs', f'epoch_{epoch:03d}')
    batches = torch.stack([torch.load(file_path) for file_path in tqdm(list(map(lambda s: osp.join(epoch_folder, s), sorted(os.listdir(epoch_folder), key=lambda s: int(s.split('_')[1])))), position=1, leave=False)])
    epochs_batches.append(batches)

storage_space = sum(map(lambda batches: sys.getsizeof(batches.storage()), epochs_batches))
print(storage_space / 1024 / 1024, 'MB')
```
